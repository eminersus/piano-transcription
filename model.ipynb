```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, TensorBoard
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import datetime



def get_image_paths(folder_path, extensions=(".jpg", ".png", ".jpeg")):
    """
    Get a list of all image file paths in a folder and its subfolders.
    Args:
        folder_path (str): Path to the folder containing images.
        extensions (tuple): File extensions to include.
    Returns:
        list: List of full image paths.
    """
    image_paths = []
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(extensions):
                image_paths.append(os.path.join(root, file))
    return image_paths

# ======================
# Data Preprocessing
# ======================
def preprocess_frame(frame, target_size=(32, 224)):
    """
    Resize the input frame to a fixed size (height, width).
    Args:
        frame (np.ndarray): Original frame as a NumPy array (H, W, C).
        target_size (tuple): Desired (height, width).
    Returns:
        np.ndarray: The resized/preprocessed frame.
    """
    resized = cv2.resize(frame, (target_size[1], target_size[0]), interpolation=cv2.INTER_AREA)
    resized = resized.astype('float32') / 255.0  # Normalize to [0, 1]
    return resized

def load_frames(paths, labels, target_size=(32, 224)):
    """
    Loads and preprocesses frames from disk into NumPy arrays.
    """
    images = []
    for p in paths:
        frame = cv2.imread(p)
        if frame is None:
            raise ValueError(f"Could not read image from path: {p}")
            
        # If grayscale, convert to 3-channel
        if len(frame.shape) == 2:
            frame = cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR)
        preprocessed = preprocess_frame(frame, target_size)
        images.append(preprocessed)

    images = np.array(images)  # shape: (n_samples, 32, 224, 3)
    labels = np.array(labels)
    return images, labels

# ======================
# Model Definition
# ======================
def build_enhanced_cnn(input_shape=(32, 224, 3), num_classes=3):
    """
    Builds a slightly deeper CNN to increase model capacity.
    """
    model = models.Sequential()

    # Convolutional Block 1
    model.add(layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))

    # Convolutional Block 2
    model.add(layers.Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Convolutional Block 3 (extra block)
    model.add(layers.Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2, 2)))

    # Flatten + Dense
    model.add(layers.Flatten())
    model.add(layers.Dense(256, activation='relu'))
    model.add(layers.Dropout(0.5))  # Additional dropout for regularization
    model.add(layers.Dense(num_classes, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model

# ======================
# Data Loading
# ======================
# Replace these with actual paths to your data
white_key_folder = "real_dataset/5/1"
black_key_folder = "real_dataset/5/2"
neither_folder = "real_dataset/5/0"

white_key_paths = get_image_paths(white_key_folder)
black_key_paths = get_image_paths(black_key_folder)
neither_paths = get_image_paths(neither_folder)

all_image_paths = white_key_paths + black_key_paths + neither_paths
all_labels = (
    [1] * len(white_key_paths)  # e.g., label 1 for white_key_pressed
    + [2] * len(black_key_paths)  # e.g., label 2 for black_key_pressed
    + [0] * len(neither_paths)    # e.g., label 0 for neither
)

# Train/Validation Split
train_paths, val_paths, train_labels, val_labels = train_test_split(
    all_image_paths,
    all_labels,
    test_size=0.2,
    shuffle=True,
    random_state=42
)
print(train_paths)
print(val_paths)
print(train_labels)
print(val_labels)

# Preprocess the frames
X_train, y_train = load_frames(train_paths, train_labels)
X_val, y_val = load_frames(val_paths, val_labels)

print("Training set shapes:", X_train.shape, y_train.shape)
print("Validation set shapes:", X_val.shape, y_val.shape)

# ======================
# Data Augmentation
# ======================
# Using tf.data or simply a Keras ImageDataGenerator-like pipeline
# Here, we'll use tf.keras.layers for on-the-fly augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomContrast(factor=0.1)           # Random zoom
    # You can add more augmentation layers if needed
])

# If your memory allows, you could convert your arrays into TF Dataset for cleaner pipeline
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
val_dataset = tf.data.Dataset.from_tensor_slices((X_val, y_val))

# Shuffle, Batch, and apply augmentation to train dataset only
batch_size = 32
train_dataset = (
    train_dataset
    .shuffle(buffer_size=len(X_train))
    .batch(batch_size)
    .map(lambda x, y: (data_augmentation(x, training=True), y), 
         num_parallel_calls=tf.data.AUTOTUNE)
    .prefetch(tf.data.AUTOTUNE)
)

val_dataset = (
    val_dataset
    .batch(batch_size)
    .prefetch(tf.data.AUTOTUNE)
)

# ======================
# Model Building
# ======================
model = build_enhanced_cnn(input_shape=(32, 224, 3), num_classes=3)

# ======================
# Callbacks for Logging and Early Stopping
# ======================
# 1. TensorBoard for real-time logs
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

# 2. EarlyStopping to stop training if validation loss doesn't improve
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,            # Number of epochs with no improvement
    restore_best_weights=True
)

# 3. ModelCheckpoint to save the best model based on validation accuracy
model_checkpoint = ModelCheckpoint(
    'best_model.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

# ======================
# Model Training
# ======================
epochs = 20

history = model.fit(
    train_dataset,
    epochs=epochs,
    validation_data=val_dataset,
    callbacks=[tensorboard_callback, early_stopping, model_checkpoint],
    verbose=1
)

# ======================
# Plot Training Curves
# ======================
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# ======================
# Save Final Model
# ======================
# Even if you have ModelCheckpoint saving the best model, 
# you can also save the final state of the model at the end if desired.
model.save("final_model.h5")
print("Final model weights saved to 'final_model.h5'.")

# ======================
# Inference
# ======================
def classify_frame(model, frame, target_size=(32, 224)):
    """
    Classify a single frame.
    Args:
        model (tf.keras.Model): Trained Keras model.
        frame (np.ndarray): Frame as a NumPy array (H, W, C).
    Returns:
        str: O
    """
    label_map = {0: 'neither', 1: 'white', 2: 'black'}
    
    # Preprocess
    if len(frame.shape) == 2:
        frame = cv2.cvtColor(frame, cv2.COLOR_GRAY2BGR)
    preprocessed = preprocess_frame(frame, target_size)
    preprocessed = np.expand_dims(preprocessed, axis=0)  # shape (1, H, W, C)
    
    # Forward pass
    preds = model.predict(preprocessed)
    pred_label = np.argmax(preds, axis=1)[0]
    return label_map[pred_label]

# Example usage
# test_frame = cv2.imread("path_to_test_frame.jpg")  # Replace with actual test frame path
# classification = classify_frame(model, test_frame)
# print("Frame classification:", classification)
```
